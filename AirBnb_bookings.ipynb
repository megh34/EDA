###Handle Missing Values
* name and host_name: Replace with placeholders.
* reviews_per_month: Fill with 0 (assuming missing values imply no reviews).
* last_review: Convert to datetime and drop rows if necessary.


````python
# Fill missing values for 'name' and 'host_name'
df['name'].fillna('No name provided', inplace=True)
df['host_name'].fillna('No host name', inplace=True)

# Fill missing 'reviews_per_month' with 0
df['reviews_per_month'].fillna(0, inplace=True)

# Convert 'last_review' to datetime and drop rows with missing values
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df = df.dropna(subset=['last_review'])

# Verify missing values
print(df.isnull().sum())
````

Remove or Cap Outliers
Remove extreme values for columns like price and minimum_nights.

````python
# Boxplots to visualize outliers
plt.figure(figsize=(12, 5))

# Boxplot for price
plt.subplot(1, 2, 1)
plt.boxplot(df['price'], vert=False)
plt.title('Boxplot of Price')
plt.xlabel('Price')

# Boxplot for minimum nights
plt.subplot(1, 2, 2)
plt.boxplot(df['minimum_nights'], vert=False)
plt.title('Boxplot of Minimum Nights')
plt.xlabel('Minimum Nights')
````

Outliers are those values which are less than 5th percentile or greater than 95th percentile.

````python
df[df['price']>=df['price'].quantile(0.95)]
````

````python
df[df['price']<df['price'].quantile(0.05)]
````

````python
# values within 5th percentile and 95th percentile
df[(df['price'] >= df['price'].quantile(0.05)) &
        (df['price'] <= df['price'].quantile(0.95))]
````

````python

# Filter out price outliers (e.g., above $1000 or below $10)
df = df[(df['price'] <= 1000) & (df['price'] >= 10)]

# Exclude listings with minimum nights greater than 365
df = df[df['minimum_nights'] <= 365]

# Display the shape of the filtered dataset and a preview
df.shape,df.head()
````

````python
print(df['price'].describe())
print(df['minimum_nights'].describe())
````


Add useful columns like year_of_last_review

````python
# Extract the year of the last review
#Extracting the year helps analyze trends in reviews over time, like identifying peak review years or periods of inactivity.
df['year_of_last_review'] = df['last_review'].dt.year
````

````python
#Estimated Revenue
#This metric estimates the revenue generated by a listing, offering insights into which properties are more profitable.
df['estimated_revenue'] = df['price'] * df['minimum_nights']
````

````python
#Availability Ratio
#This indicates the proportion of the year a listing is available, helping identify listings with high or low availability.
df['availability_ratio'] = df['availability_365'] / 365
````

````python

#Month of Last Review
#Extracting the month provides seasonality insights, such as which months see the most reviews.
df['month_of_last_review'] = df['last_review'].dt.month
````

````python
# Categorize prices into low, medium, and high
#Categorizing prices simplifies the analysis of affordability and market segmentation.
df['price_category'] = 'low'
df.loc[df['price'] > 50, 'price_category'] = 'medium'
df.loc[df['price'] > 200, 'price_category'] = 'high'
````

### What all manipulations have you done and insights you found?

The dataset underwent several data cleaning and preprocessing steps. Missing values in the name and host_name columns were replaced with placeholder text, while reviews_per_month was filled with 0. Outliers in price were filtered out, and listings with minimum_nights greater than 365 were excluded. New features were engineered, including year_of_last_review, estimated_revenue, availability_ratio, month_of_last_review, and price_category (low, medium, high). Aggregated metrics, such as average price by neighborhood, median revenue by room type, and review trends by year and month, were calculated. Insights revealed that revenue is weakly correlated with the number of reviews, availability ratios vary by room type, and certain neighborhoods and room types dominate in revenue generation. Seasonal trends in reviews and listings with high availability were identified, offering valuable insights into market segmentation, profitability, and customer behavior.

High Prices: Listings with prices above 200 are categorized as "high" and predominantly found in neighborhoods like Manhattan and luxury areas. These listings include entire apartments and premium properties.
Low Prices: Listings priced below 50, categorized as "low," are concentrated in less central areas such as Queens and Brooklyn. These tend to be private or shared rooms.
Medium Prices: The majority of listings fall in the "medium" price range (50-200), providing a balance of affordability and comfort.

High Revenue: "Entire home/apartment" listings generate the highest median estimated revenue (480), suggesting they are the most profitable category.
Low Revenue: "Shared rooms" generate the lowest revenue (76), likely due to their affordability and limited appeal to travelers seeking privacy.
